```python
from utils import ffr, trajectory_dt, Helpers
import plotly.express as px
import plotly.graph_objects as go
from datetime import datetime
import pandas as pd 
```

```python
apr23 = ffr(4, 23, '2022-10-01')
mar23 = ffr(3, 23, '2022-10-01')
feb23 = ffr(2, 23, '2022-10-01')
jan23 = ffr(1, 23, '2022-10-01')
dec23 = ffr(12, 22, '2022-10-01')
```

```python
fig = go.Figure()

dfs = {'dec23':dec23.futures_prices, 'jan23':jan23.futures_prices, 
       'feb23':feb23.futures_prices, 'mar23':mar23.futures_prices,
       'apr23':apr23.futures_prices}

for i in dfs:
    fig.add_trace(go.Scatter(x  = dfs[i]['DT'], y = dfs[i]['R_IMPL'], name = i))

fig.show()
```

```python
dt = datetime.strptime("2022-11-01", '%Y-%m-%d' )
```

```python
df = pd.DataFrame(columns = ['dt', '2022-12-01', '2023-01-01', '2023-02-01', '2023-03-01', '2023-04-01', '2023-05-01', '2023-06-01', '2023-07-01'])
for i in [0, 7, 14, 21, 30, 45, 60]:

    curdt = Helpers.days_n_before(dt, i)

    obj = trajectory_dt(curdt).price()
    df = df.append(obj,ignore_index=True)
df.columns = ['Date', 'Dec22', 'Jan23', 'Feb23', 'Mar23', 'Apr23', 'May23', 'Jun23', "Jul23"]
```

```python
df.head()
```

```python
fig = go.Figure()
for proj_month in df.columns[1:]:

    fig.add_trace(go.Scatter(x = df['Date'], y = df[proj_month], name = proj_month))

fig.show()
```
