```python
from utils import ffr, trajectory_dt, Helpers, ffr_months_df
import plotly.express as px
import plotly.graph_objects as go
from datetime import datetime
import pandas as pd 
```

```python
df = ffr_months_df().df

dec23 = df[['Date', '2023-12-01']]
jun23  = df[['Date', '2023-06-01']]
apr23 = df[['Date', '2023-04-01']]
feb23 = df[['Date', '2023-02-01']]
dec22 = df[['Date', '2022-12-01']]
```

```python
fig = go.Figure()
fig.update_layout(title = "Implied FFR at certain month")
dfs = {'dec22':dec22, 
       'feb23':feb23,
       'apr23':apr23, 
       'jun23':jun23,
       'dec23':dec23}

for i in dfs:
    dfpp= dfs[i]
    cols = dfpp.columns
    fig.add_trace(go.Scatter(x = df[cols[0]], y = df[cols[1]], name = i))

fig.show()
```

```python
dt = datetime.strptime("2022-11-01", '%Y-%m-%d' )
```

```python
df = pd.DataFrame(columns = ['dt', '2022-12-01', '2023-01-01', '2023-02-01', '2023-03-01', '2023-04-01', '2023-05-01', '2023-06-01', '2023-07-01'])
for i in [0, 7, 14, 21, 30, 45, 60]:

    curdt = Helpers.days_n_before(dt, i)
    obj = trajectory_dt(curdt).price()
    df = df.append(obj,ignore_index=True)
df.columns = ['Date', 'Dec22', 'Jan23', 'Feb23', 'Mar23', 'Apr23', 'May23', 'Jun23', "Jul23"]
```

```python
fig = go.Figure()
for proj_month in df.columns[1:]:

    fig.add_trace(go.Scatter(x = df['Date'], y = df[proj_month], name = proj_month))

fig.show()
```
