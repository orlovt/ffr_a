```python
from utils import  FFR__df
import plotly.graph_objects as go
from datetime import datetime
```

```python
#using "FFR__df" class to parse, clean and transform Fed Funds Futures data

df = FFR__df().df

#plotting futures data
#0y axis is the rate 
#0x date 
#legends correspond to futures contracts from yahoo finance  

fig_2 = go.Figure()
fig_2.update_layout(title = "Implired FedFundsRate for given month")

#setting number of months and intervals 
for i in df.columns[1::3]: 
    #converting column name into datetime format, for lines naming 
    dt = datetime.strptime(i, '%Y-%m-%d' )
    fig_2.add_trace(go.Scatter(x = df['Date'], y = df[i], name = dt.strftime("%b") + dt.strftime("%y")))
fig_2.show()
```
